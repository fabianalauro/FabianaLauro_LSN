## Exercise 05

The aim of this exercise is to use the Metropolis algorithm to sample the probability distribution of the $1s$ and $2p$ orbital of an hydrogen atom, in order to estimate $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ and $\left\langle r \right\rangle_{\Psi_{2,1,0}}$.

I worked in **Cartesian coordinates** and using **Bohr radius units** for distances (rescaling $r=r/a_0$). In this way, the wave functions of the system become:
- $\Psi_{1,0,0}(x, y, z)= {1\over{\sqrt{\pi}}} e^{-\sqrt{x^2+y^2+z^2}}$ for the ground state
- $\Psi_{2,1,0}(x, y, z)= \sqrt{{2\over\pi}} {z\over 8} e^{-{\sqrt{x^2+y^2+z^2}\over 2}}$ for the first excited state

Thus the probability distibutions become:
- $|\Psi_{1,0,0}(x, y, z)|^2= {1\over\pi} e^{-2\sqrt{x^2+y^2+z^2}}$ for the ground state
- $|\Psi_{2,1,0}(x, y, z)|^2=  {z^2\over 32 \pi} e^{-\sqrt{x^2+y^2+z^2}}$ for the first excited state

And the expectation values of the radius $r$ are: 
- $\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}$ for the ground state
- $\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 $ for the first excited state


I sampled the **probability distributions** reported above through the **Metropolis algorithm**, using two different transition probablilities $T(\vec{x}|\vec{y})$: the **uniform** one and the **multivariate normal** one (i.e. a Gaussian for ech coordinate, from now on I'll refer to it as *GAUSS* transition probability). So I made in total four simulations.

#### Equilibration

I started each simulation with the **equilibration** process: since the Metropolis algorithm tends to the target probability distribution only asimptotically, I decided to choose the point $(x, y, z)=(20, 20, 20)$ as the starting point of my Markov process (a point with low probability) and to make $10^3$ **equilibration steps** to get to a high probability area before starting the actual simulation.

To decide the **size of the step** taken, I proceeded by trials and errors until the acceptance rate $\alpha$ of the Metropolis algorithm was close to $50\%$. The values I used in the end are:
- $\Psi_{1, 0, 0}$, *UNIFORM*: $\delta=1.3$ ($\alpha = 0.504$)
- $\Psi_{1, 0, 0}$, *GAUSS*:   $\sigma=0.8$ ($\alpha = 0.506$)
- $\Psi_{2, 1, 0}$, *UNIFORM*: $\delta=2.9$ ($\alpha = 0.500$)
- $\Psi_{2, 1, 0}$, *GAUSS*:   $\delta=1.9$ ($\alpha = 0.501$)


I plotted the points sampled during the equilibration process on the 3D graphs reported below, they show that $10^3$ equilibration steps are way sufficient to get to a high probability area.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

Ux, Uy, Uz = np.loadtxt ("FFO_100.dat", unpack = "true", usecols = (0, 1, 2) ) 
Gx, Gy, Gz = np.loadtxt ("FFO_100.dat", unpack = "true", usecols = (3, 4, 5) ) 

fig = plt.figure(figsize = (17, 9))
plt.suptitle ("Equilibration path ($\psi_{1,0,0}$)", fontsize = 30)

ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.scatter(Ux, Uy, Uz, c=Ux, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("UNIFORM")
ax.view_init(10, 20)

ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.scatter(Gx, Gy, Gz, c=Gx, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("GAUSS")
ax.view_init(10, 20)

Ux, Uy, Uz = np.loadtxt ("FFO_210.dat", unpack = "true", usecols = (0, 1, 2) ) 
Gx, Gy, Gz = np.loadtxt ("FFO_210.dat", unpack = "true", usecols = (3, 4, 5) ) 

fig = plt.figure(figsize = (17, 9))
plt.suptitle ("Equilibration path ($\psi_{2,1,0}$)", fontsize = 30)

ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.scatter(Ux, Uy, Uz, c=Ux, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("UNIFORM")
ax.view_init(10, 20)

ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.scatter(Gx, Gy, Gz, c=Gx, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("GAUSS")
ax.view_init(10, 20)

plt.show()

```

#### Simulation

After the equilibration, I started the actual **simulation**.
For each simulation I sampled $M=10^6$ steps and I divided them in $N=100$ blocks to calculate $\langle r \rangle$ using the **blocking method**.
I also took track of the sampled points, recording one position every 100 steps.

To decide the **size of the step** taken, I proceeded as explained above (in the *equilibration* part) and the values I used in the end are:
- $\Psi_{1, 0, 0}$, *UNIFORM*: $\delta=1.2$ ($\alpha = 0.508$)
- $\Psi_{1, 0, 0}$, *GAUSS*:   $\sigma=0.8$ ($\alpha = 0.482$)
- $\Psi_{2, 1, 0}$, *UNIFORM*: $\delta=2.8$ ($\alpha = 0.523$)
- $\Psi_{2, 1, 0}$, *GAUSS*:   $\delta=1.9$ ($\alpha = 0.496$)

Below I reported the graphs showing the average values of $r$ as a function of the number of the blocks and the 3D graphs displaying the points sampled during the simulation (one every 100 steps).
The graphs show a good compatiblity with the expected results and point out there is no significative difference between the results obtained using uniform distribution the ones obtained using normal distribution.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

#Loading from file
Ux, Uy, Uerror = np.loadtxt ("r_100U.dat", unpack = "true" ) #, usecols = (0, 1,2)
Gx, Gy, Gerror = np.loadtxt ("r_100G.dat", unpack = "true" ) #, usecols = (0, 3,4)
#Plotting graph
figure = plt.figure (figsize=(13, 8))

plt.errorbar (Ux, Uy, yerr = Uerror, label = "Uniform", color = "royalblue")
plt.errorbar (Gx, Gy, yerr = Gerror, label = "Gauss", color = "darkorange")
plt.hlines (3/2, 0, 100, color = "black", linestyles = "dashed")
plt.xlabel("N", fontsize = 15)
plt.ylabel("$<r>$", fontsize = 15)
plt.title("Estimation of $< r > _{\psi_{1,0,0}}$", fontsize = 40)
plt.grid("true")
plt.legend()

#Loading from file
Ux, Uy, Uerror = np.loadtxt ("r_210U.dat", unpack = "true" ) #, usecols = (0, 1,2)
Gx, Gy, Gerror = np.loadtxt ("r_210G.dat", unpack = "true" ) #, usecols = (0, 3,4)
#Plotting graph
figure = plt.figure (figsize=(13, 8))

plt.errorbar (Ux, Uy, yerr = Uerror, label = "Uniform", color = "royalblue")
plt.errorbar (Gx, Gy, yerr = Gerror, label = "Gauss", color = "darkorange")
plt.hlines (5, 0, 100, color = "black", linestyles = "dashed")
plt.xlabel("N", fontsize = 15)
plt.ylabel("$<r>$", fontsize = 15)
plt.title("Estimation of $< r > _{\psi_{2,1,0}}$", fontsize = 40)
plt.grid("true")
plt.legend()

plt.show()
```

#### Notes on the code

During the implementation of the code, I made use of some **classes** I created.

To perform the **Metropolis algorithm**, I created a class called <font face="Courier">Metropolis3D</font> which takes advantage of the polymorphism of the virtual class <font face="Courier">Basic_PDF</font> to accept as data member the probability distribution wanted, implemented as a derived class of <font face="Courier">Basic_PDF</font>.
It has also, as data members, three doubles which represent the coordinates of the current position, as well as a double to record the acceptance rate of every step, and a pointer to a random number generator.
It has then two methods (<font face="Courier">MRT2_Unif</font> and <font face="Courier">MRT2_Gauss</font>) which make a step with the Metropolis algorithm using respectively che uniform transition probability and the multivariate normal transition probability.

To compute the **blocking average**, then, I implemented a class called <font face="Courier">Blocking</font>, which, through a method called <font face="Courier">Add</font>, performs the blocking average of the added values and writes the results on a file.
