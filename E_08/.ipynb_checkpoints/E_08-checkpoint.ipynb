## Exercise 08

The aim of this exercise is to estimate the **ground state energy** and the **ground state wave function** of a single **quantum particle** in a one dimensional (1D) space.
The external potential in which the particle is confined is: 
$$
V(x) = x^4 - \frac{5}{2}x^2
$$
So the Hamiltonian of the system, in the coordinate representation, is:
$$
H(x) = -\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2} + V(x)
$$

The method used for the optimisation is the **Variational Monte Carlo (VMC)** technique, with trial wave function 

$$\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+e^{-\frac{(x+\mu)^2}{2\sigma^2}}$$ 
I had to optimise the two parameters $\sigma$ and $\mu$ in order to find the **minimum** value of $\langle {\hat H} \rangle_T = \frac{\langle \Psi_T^{\sigma,\mu}| {\hat H} | \Psi_T^{\sigma,\mu} \rangle}{\langle \Psi_T^{\sigma,\mu} | \Psi_T^{\sigma,\mu} \rangle}$, which is a good upper-bound approximation of the ground state energy $E_0 = \frac{\langle \Psi_0| {\hat H} | \Psi_0 \rangle} {\langle \Psi_0 | \Psi_0 \rangle}$.

## Exercise 08.1

Given a wave function $\Psi_T(x)$ , the **expectation value** of the Hamiltonian (average value of energy) is calculated as:

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)} =
\int dx p(x) E_{loc}(x)
$$
where 
$$p(x) = \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2}$$ 
and 
$$
E_{loc}(x) = \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)} =
\frac{\left[{ -\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2}} + V(x) \right] \Psi_T^{\sigma,\mu}(x)}{\Psi_T^{\sigma,\mu}(x)} =
-\frac{x^2+\mu^2-\sigma^2-2\mu x \tanh(\frac{\mu x}{\sigma^2})}{2\sigma^4}+x^4-\frac{5}{2}x^2
$$

Through the **Metropolis algorithm** I was able to sample the probability distribution $p(x)$, and use the values obatined to calculate $E_{loc}(x)$. 
Repeating the procedure multiple times, for fixed values of $\mu$ and $\sigma$, and calculating the average value of $E_{loc}(x)$ I coud get $\langle {\hat H} \rangle_T$ for that values of $\mu$ and $\sigma$.

```python

pos = np.loadtxt("Positions.dat")

def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2
    
hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)


# Plot a few things
plt.figure(figsize=(15,8))
scale = 0.3
#plt.plot(x, scale*V, color="Black", label="Potential") # plot the potential
plt.hist (pos, 50, density = True, histtype = "step", label="VMC solution")
plt.plot(x,(psi[0])**2, label="Analytical solution")
plt.title("Ground state wave function")
plt.xlabel("x")
plt.ylabel("$\psi$")

plt.grid(True)
plt.xlim((-3,3))
#plt.ylim((-0.6,0.6))
plt.legend()
plt.show()
```

The first graph shows that, although the value of the ground state energy found is not compatible with the one calculated analytically, it is a good upper-bound approximation of it, being only 0.029146 units of energy apart.

The first graph shows that the ground state wave functions found with the two methods are quite compatible in shape, with the discrepances in proximity of the peaks of the function.
