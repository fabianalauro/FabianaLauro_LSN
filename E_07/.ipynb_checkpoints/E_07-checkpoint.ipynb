The aim of this exercise is to modify a given code to perform a simulation of a Lennard-Jones fluid in a canonical ensemble (NVT), using the Metropolis algorithm, to study the evolution of two abservables: **potential energy per particle**, $U/N$, and **pressure**, $P$

I worked with **Argon**, its caratheristic values in Lennard-Jones units are:

- $\epsilon/k_B = 120$ K (temperature unit)
- $m=39.948$ amu (mass unit)

These are the parameters I used to simulate the three phases:

## Exercise 07.1

This part of the exercise is intended to study the **autocorrelation** of the data and understand how large the **blocks**, needed to calculate the averages through the data blocking technique, need to be.
To accomplish this task I had to write some Python scripts using of the $5x10^5$ **instant data** produced by the first run of the program.

#### Autocorrelation

In the python scripts below I calculated the autocorrelation of the data through the formula: 

$$ Ac_{[x]}(\tau) = \frac{\langle x(t)x(t+\tau) \rangle - \langle x(t)\rangle \langle x(t+\tau) \rangle}{\sigma^2_x} $$

for 50 different values of $\tau$ (spaced equally from $\tau = 10$ to $\tau = 500$).

Below I reported the plots of the autocorrelation, as a function of $\tau$, for every phase. 

```python
#SOLID: Calculating autocorrelation

S_epot, S_press = np.loadtxt("output.instant.S", unpack = "true", skiprows=(1001))

M = 5 * (10**5)
xA = [i*10 for i in range (1, 51)]

print("it may take a while...")

#EPOT
#calculating variance
sum = 0
sum2 = 0
sigma = 0

for i in range (0, M):
    sum += S_epot [i]
    sum2 += S_epot [i]*S_epot [i]
sigma = sum2/M - (sum/M)**2

#calculating autocorrelation
k = 10
S_AcU = [0] * 50 #autocorrelation
while (k <= 500) :
    prod = 0
    xoft = 0
    count = 0
    xoftptau = 0
    for i in range(0, M-k):
        prod += S_epot[i] * S_epot[i+k]
        xoft += S_epot[i]
        xoftptau += S_epot[i+k]
        count = count + 1
    corr = (prod/count - xoft*xoftptau/(count*count)) / sigma
    S_AcU[int(k/10-1)] = corr
    k += 10
    
#PRESS
#calculating variance
sum = 0
sum2 = 0
sigma = 0

for i in range (0, M):
    sum += S_press [i]
    sum2 += S_press [i]*S_press [i]
sigma = sum2/M - pow (sum/M, 2) 

#calculating autocorrelation
k = 10
S_AcP = [0] * 50 #autocorrelation
while (k <= 500) :
    prod = 0
    xoft = 0
    count = 0
    xoftptau = 0
    for i in range(0, M-k):
        prod += S_press[i] * S_press[i+k]
        xoft += S_press[i]
        xoftptau += S_press[i+k]
        count = count + 1
    corr = (prod/count - xoft*xoftptau/(count*count)) / sigma
    S_AcP[int(k/10-1)] = corr
    k += 10    
print ("Done!")    
```

```python
#Plotting it

fig = plt.figure (figsize = (16, 6))
plt.suptitle ("SOLID: Autocorrelation", fontsize = 25)
    
#EPOT 
plt.subplot (1, 2, 1)
plt.plot (xA, S_AcU, color = "red")

plt.title("Potential energy", fontsize=15)
plt.xlabel ("Length of the steps")
plt.ylabel ("Autocorrelation")
plt.grid()
    
#PRESS 
plt.subplot (1, 2, 2)
plt.plot (xA, S_AcP, color = "red")

plt.title("Pressure", fontsize=15)
plt.xlabel ("Length of the steps")
plt.ylabel ("Autocorrelation")
plt.grid()

plt.show()
```
