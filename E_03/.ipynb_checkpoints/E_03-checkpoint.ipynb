The aim of this exercise is to use Monte Carlo methods to calculate the right price of a **european option** in a **financial market**.

This simulation is based on the theory that the evolution of the price of an asset can be statiscally calculated using the model of a **Geometric Brownian Motion (GBM)**. 
In this way, given some parameters (the asset price at the initial time $S(0)$, the volatility of the market $\sigma$, the risk-free interest rate $ùëü$), one can calculate the asset price at the final time $S(T)$.

Once calculated $S(T)$, the profit of the holder of the option can be easily found through the formulas: $profit=max[0,S(T)-K]$ for a **call option**; $profit=max[0,K-S(T)]$ for a **put option**, where $K$ is the strike price of the option.

Making a large number ($M \ge 10^4$) of simulations, the average profit of the holder, divided by a factor $\exp(-rT)$, corresponds to the **option price** at the initial time $t=0$, defined as $C[S(0),0]$ for a call and $P[S(0),0]$ for a put.

I made four estimations of the option price, two for a Call-Option and two for a Put-Option, using two different ways of calculating the final asset price:

- The first way consists in sampling **directly** the final asset price (**\"Direct method\"**), through the formula: 
$$S(T) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)T+\sigma W(T)\right]$$ where $W(t)$ is a random number sampled according to a normal distribution, with $\mu = 0$, $\sigma = t$.

- The second estimation was made dividing $[0, T]$ into 100 time intervals and calculating the discretised path of the stocastic process $S(t)$ (**\"Discretised method\"**) in a recursive way, through the formula 
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    where $Z_i$ is a normally distributed random number, with $\mu = 0$, $\sigma = 1$ .

Every estimation was made $M=10^5$ times, I then computed the average using the **blocking method** (with $N = 100$).


In order to write a slimmer and more readable main code I created the class **'PlainVanilla'** in which I implemented all the class methods necessary to estimate the Call and Put options using both the direct and discretised methods.

The **parameters** I used for the simulation are the following:
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$


Below I plotted the values I got for $C[S(0),0]$ and $P[S(0),0]$ using the two methods, as a function of the number of the blocks $N$. I also compared the results with the analytical solutions for the price, given by the Black-Scholes equation (the dashed line in the plots).

```python
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt

#Loading from file
CDx, CDy, CDerror = np.loadtxt ("Results_dir.dat", usecols = (0, 1,2), unpack = "true")
PDx, PDy, PDerror = np.loadtxt ("Results_dir.dat", usecols = (0, 3,4), unpack = "true")
Cx, Cy, Cerror = np.loadtxt ("Results_dir.dat", usecols = (0, 1,2), unpack = "true")
Px, Py, Perror = np.loadtxt ("Results_discr.dat", usecols = (0, 3,4), unpack = "true")

figure = plt.figure (figsize=(15, 10))
plt.suptitle ("Plain Vanilla option pricing")

#Graph 1: call, direct
plt.subplot(2, 2, 1)
plt.errorbar (CDx, CDy, yerr = CDerror)
plt.hlines (14.975790778311286, 0, 100, color = "black", linestyles="dashed")
plt.grid(True)
plt.title ("Call-option price (simulated with \"Direct\" method)")
plt.xlabel ("N")
plt.ylabel ("C [S(0),0]")
#plt.legend()

#Graph 2: put, direct
plt.subplot(2, 2, 2)
plt.errorbar (PDx, PDy, yerr = PDerror)
plt.hlines (5.4595325819072364, 0, 100, color = "black", linestyles="dashed")
plt.grid(True)
plt.title ("Put-option price (simulated with \"Direct\" method)")
plt.xlabel ("N")
plt.ylabel ("P [S(0),0]")
#plt.legend()

#Graph 3: call, discretised
plt.subplot(2, 2, 3)
plt.errorbar (Cx, Cy, yerr = Cerror)
plt.hlines (14.975790778311286, 0, 100, color = "black", linestyles="dashed")
plt.grid(True)
plt.title ("Call-option price (simulated with \"Discretised\" method)")
plt.xlabel ("N")
plt.ylabel ("C [S(0),0]")
#plt.legend()

#Graph 4: put, discretised
plt.subplot(2, 2, 4)
plt.errorbar (Px, Py, yerr = Perror)
plt.hlines (5.4595325819072364, 0, 100, color = "black", linestyles="dashed")
plt.grid(True)
plt.title ("Put-option price (simulated with \"Discretised\" method)")
plt.xlabel ("N")
plt.ylabel ("P [S(0),0]")
#plt.legend()


plt.show()

#print (Py)
```

All the pictures converge well to the analytical solution. 
This is due to the fact that the method used to calculate $S(t)$ is an **exact method**, so it gives the same results of the analytical method.
