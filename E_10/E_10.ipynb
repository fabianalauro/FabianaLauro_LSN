## Exercise 10.1

This exercise asked to adapt our genetic algorithm to solve the TSP with a **Simulated Annealing** (SA) algorithm.
Also in this case, the algorithm had to be applied to both 32 cities on a circumference and 32 cities inside a square.

The idea behind a SA algorithm is to create an imaginary thermodynamic system with, as **energy function**, the function to minimise.
The **temperature** of the system is lowered extremely slowly, and for every change in the temperature a new configuration is sampled, through the **Metropolis algorithm**, with the **Boltzmann weight** ($P=\frac{e^{-\beta L}}{Z}$) as probability distribution.

In this case, the **length of the trip** represented the energy of the system.
I started with a temperature $T_i=10$ and I multiplied it by $0.99999$ until it reached the value $T_f=0.01$.

To do so, I created a class called `Annealing` which performed the sampling via Metropolis Algorithm and took track of the best trip found.
